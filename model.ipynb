<h1>Exploratory Data Analysis</h1>

```python
import pandas as pd     #import pandas
```

```python
df = pd.read_csv('laptop_price.csv', encoding='latin-1')        #load dataset
df.head()
```

```python
df.info()
```

```python
df['Weight'] = df['Weight'].astype(str).str.replace('kg', '').astype(float)     #remove 'kg' from 'Weight' column
df['Ram'] = df['Ram'].astype(str).str.replace('GB', '').astype(int)     #remove 'GB' from 'Ram' column
df = df.drop(columns=['laptop_ID', 'Product'], axis=1)      #drop columns
df.head()
```

```python
df['Company'].value_counts()
```

```python
def company(x):     
    if x == 'Dell' or x == 'Lenovo' or x == 'HP' or x == 'Asus' or x == 'MSI' or x == 'Acer' or x == 'Toshiba' or x == 'Apple' or x == 'Samsung':
        return x
    else:
        return 'Other'
    
df['Company'] = df['Company'].apply(company)        #update 'Company' column
df['Company'].value_counts()
```

```python
def split_resolution(x):
    return x.split(' ')[-1]

def pop_resolution(x):
    y = x.split(' ')
    y.pop()
    return ' '.join(y)

df['Resolution'] = df['ScreenResolution'].apply(split_resolution)       #add new column
df['ScreenResolution'] =df['ScreenResolution'].apply(pop_resolution)        #update 'ScreenResolution' column
df.head()
```

```python
df['ScreenResolution'].value_counts()
```

```python
def full_hd(x):
    if 'Full HD' in x:
        return 1
    else:
        return 0 

def ips_panel(x):
    if 'IPS Panel' in x:
        return 1
    else:
        return 0

def ultra_hd(x):
    if '4K Ultra HD' in x:
        return 1
    else:
        return 0

def touch_screen(x):
    if 'Touchscreen' in x:
        return 1
    else:
        return 0

def quad_hd(x):
    if 'Quad HD+' in x:
        return 1
    else:
        return 0

def ratina_display(x):
    if 'Retina Display'in x:
        return 1
    else:
        return 0

#make new columns from 'ScreenResolution' column

df['FullHD'] = df['ScreenResolution'].apply(full_hd)
df['IpsPanel'] = df['ScreenResolution'].apply(ips_panel)
df['UltraHD'] = df['ScreenResolution'].apply(ultra_hd)
df['TouchScreen'] = df['ScreenResolution'].apply(touch_screen)
df['QuadHD'] = df['ScreenResolution'].apply(quad_hd)
df['RatinaDisplay'] = df['ScreenResolution'].apply(ratina_display)    

df.head()
```

```python
df = df.drop('ScreenResolution', axis=1)        #drop column
df.head()
```

```python
def cpu_brand(x):
    return x.split(' ')[-1].replace('GHz','')

def pop_cpu(x):
    return x.split(' ')[0]

df['CpuSpeed'] =  df['Cpu'].apply(cpu_brand).astype(float)      #remove 'GHz' from 'Cpu' column and convert column to float()

df['Cpu'] = df['Cpu'].apply(pop_cpu)        #update 'Cpu' column
df.head()
```

```python
df['Cpu'].value_counts()
```

```python
df = df.loc[df['Cpu'] != 'Samsung']
df['Cpu'].value_counts()       #update 'Cpu' column
```

```python
def gpu(x):
    y = x.split()[:2]
    return ' '.join(y)

df['Gpu'] = df['Gpu'].apply(gpu)

def gpu_other(x):
    if x == 'Intel Graphics' or x == 'AMD R17M-M1-70' or x == 'Nvidia GTX' or x == 'AMD R4' or x == 'AMD FirePro' or x == 'ARM Mali':
        return 'Other'
    else:
        return x
    
df['Gpu'] = df['Gpu'].apply(gpu_other)
df['Gpu'].value_counts()
```

```python
df = df.drop('Gpu', axis=1)
df.head()
```

```python
df.info()
```

```python
data = pd.get_dummies(df)       #convert columns to dummies
data.info()
```

<h1>Model Selection</h1>

```python
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.linear_model import ElasticNet, SGDRegressor, LinearRegression, Lasso, BayesianRidge
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, HistGradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.kernel_ridge import KernelRidge
from xgboost.sklearn import XGBRegressor
```

```python
#define x data and y data

x = data.drop('Price_euros', axis=1)        
y = data['Price_euros']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3)     #data splitting
x_train.shape, x_test.shape
```

```python
def model_acc(model):
    model.fit(x_train,y_train)
    acc = model.score(x_test,y_test)
    cross_val = cross_val_score(model, x, y, cv=5).mean()
    print('model name == '+ str(model) + ', model accuracy = '+ str(acc) + ', corss validation = '+ str(cross_val))
```

```python
ElasticNet = ElasticNet()
SGDRegressor = SGDRegressor()
LinearRegression = LinearRegression()
Lasso = Lasso()
BayesianRidge = BayesianRidge()

model_acc(ElasticNet)
model_acc(SGDRegressor)
model_acc(LinearRegression)
model_acc(Lasso)
model_acc(BayesianRidge)
```

```python
RandomForestRegressor = RandomForestRegressor()
GradientBoostingRegressor = GradientBoostingRegressor()
HistGradientBoostingRegressor = HistGradientBoostingRegressor()

model_acc(RandomForestRegressor)
model_acc(GradientBoostingRegressor)
model_acc(HistGradientBoostingRegressor)
```

```python
SVR = SVR()
KernelRidge = KernelRidge()

model_acc(SVR)
model_acc(KernelRidge)
```

```python
XGBRegressor = XGBRegressor()

model_acc(XGBRegressor)
```

<h1>Hyperparameter Tuning</h1>

```python
param_grid = {'loss':['squared_error', 'absolute_error', 'huber', 'quantile'], 'learning_rate':[0.0, 0.01, 0.1, 0.2, 0.3], 'n_estimators': [200, 300, 400]}

grid_obj = GridSearchCV(estimator=GradientBoostingRegressor, param_grid=param_grid, cv=5, verbose=2)
grid_fit = grid_obj.fit(x_train, y_train)
best_model = grid_fit.best_estimator_
best_model.score(x_test, y_test)
```
